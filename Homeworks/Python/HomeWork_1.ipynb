### 8. Write a function which will be able to show your system configuration.

```python
def show_system_config():
    
    """ This is a function that will show us our system configuration
        This function generally shows us our cpu_name, ram_size, storage_size, gpu_name, gpu_size"""
    
    from tensorflow.python.client import device_lib
    import subprocess as sp
    import os
    import cpuinfo
    import psutil
    import shutil
    
    cpu_name = cpuinfo.get_cpu_info()['brand_raw']
    gpu_name = device_lib.list_local_devices()[1].physical_device_desc[17:39]
    
    command = "nvidia-smi --query-gpu=memory.free --format=csv"
    memory_free_info = sp.check_output(command.split()).decode('ascii').split('\n')[:-1][1:]
    memory_free_values = [int(x.split()[0]) for i, x in enumerate(memory_free_info)]
    gpu_memory = round(memory_free_values[0]/1000)
    
    total_ram = round(psutil.virtual_memory().total / 1073741824)
    
    total_ssd, used_ssd, free_ssd = shutil.disk_usage("/")
    total_hdd, used_hdd, free_hdd = shutil.disk_usage("E:/")
    
    print(f"CPU name: {cpu_name}")
    print(f"GPU name: {gpu_name}")
    print(f"GPU memory: {gpu_memory} GiB")
    print(f"Total Ram: {total_ram} GiB")
    print(f"Total SSD: {(total_ssd // (2**30))} GiB, Used SSD: {(used_ssd // (2**30))} GiB, Free SSD: {(free_ssd // (2**30))} GiB")
    print(f"Total HDD: {(total_hdd // (2**30))} GiB, Used HDD: {(used_hdd // (2**30))} GiB, Free HDD: {(free_hdd // (2**30))} GiB")
    print(f"Total Storage: {(total_hdd+total_ssd) // (2**30) / 1024} TB")
    
    
    
    
    
```

```python
show_system_config()
```

### 11. Write a function which can read video file and play for you.

```python
def video_player(x):
    """The video file must be in the current working directory"""
    
    import cv2
    import numpy as np

    cap = cv2.VideoCapture(x)

    if (cap.isOpened()==False):
        print("Error opening video file")

    while (cap.isOpened()):
        ret, frame = cap.read()
        if ret == True:
            cv2.imshow('Frame', frame)
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
        else:
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
video_player("test_video.mp4")
```

### 14. Write a function which will be able to access your mail.

```python
def email_reader(un, psw):
    
    import imaplib
    import email
    from email.header import decode_header
    
    username = un
    password = psw
    
    imap = imaplib.IMAP4_SSL("imap.gmail.com")
    imap.login(username, password)
    
    status, messages = imap.select("INBOX")
    messages = int(messages[0])
    
    for i in range(messages, 0, -1):
        res, msg = imap.fetch(str(i), "(RFC822)")
        for response in msg:
            if isinstance(response, tuple):
                msg = email.message_from_bytes(response[1])
                subject, encoding = decode_header(msg["Subject"])[0]
                if isinstance(subject, bytes):
                    subject = subject.decode(encoding)
                From, encoding = decode_header(msg.get("From"))[0]
                if isinstance(From, bytes):
                    From = From.decode(encoding)
                print("Subject:", subject)
                print("From:", From)
            
                if msg.is_multipart():
                    for part in msg.walk():
                        content_type = part.get_content_type()
                        content_disposition = str(part.get("Content-Disposition"))
                        try:
                            body = part.get_payload(decode=True).decode()
                        except:
                            pass
                        if content_type == "text/plain":
                            print(body)
                            
                else:
                    content_type = msg.get_content_type()
                    body = msg.get_payload(decode=True).decode()
                    if content_type == "text/plain":
                        print(body)
                    
                print("="*100)
            
    imap.close()
    imap.logout()
```

### 16. Write a function to read a complete PDF file.

```python
def read_pdf(filePath):

    import PyPDF2

    pdfFileObj = open(filePath, "rb")
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    pdfNumPages = pdfReader.numPages

    for i in range(0, pdfNumPages):
        pageObj = pdfReader.getPage(i)
        print(pageObj.extractText())
    
    pdfFileObj.close()
```

```python
read_pdf("data/thinkstats2.pdf")
```

### 20. Write a function by which you will be able to append two PDF files.

```python
def merge_pdf(pdf_path_1, pdf_path_2, result_pdf):
    
    from PyPDF2 import PdfFileMerger
    pdfs = [pdf_path_1, pdf_path_2]
    merger = PdfFileMerger()
    
    for pdf in pdfs:
        merger.append(pdf)

    merger.write(result_pdf)
    merger.close()
```

```python
merge_pdf('data/sample.pdf', 'data/Statistics.pdf', 'data/result.pdf')
```
